# yourgoodpants.csv





```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/yourgoodpants.csv', sep=',')
print(df)
```

```python
df.loc[df.Time_s == "--undefined--", "Time_s"] = None
df.loc[df.F0_Hz == "--undefined--", "F0_Hz"] = None
print(df)
```

```python
time = df.Time_s
pitch = df.F0_Hz
import matplotlib.pyplot as plt

plt.plot(time, pitch, 'bo', label='Pitch (Hz)')
plt.legend(loc='best', ncol=2, fontsize=14, frameon=True, shadow=True)
plt.xlabel('Time(s)')
plt.ylabel('Pitch(Hz)')

plt.show()
```

# yourgoodpants_formants

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/yourgoodpants_formants.csv', sep=',')
print(df)
```

```python
df.loc[df.Time_s == "--undefined--", "Time_s"] = None
df.loc[df.F1_Hz == "--undefined--", "F1_Hz"] = None
df.loc[df.F2_Hz == "--undefined--", "F2_Hz"] = None
df.loc[df.F3_Hz == "--undefined--", "F3_Hz"] = None
df.loc[df.F4_Hz == "--undefined--", "F4_Hz"] = None
print(df)
```

```python
import matplotlib.pyplot as plt

time = df.Time_s
plt.xlabel('Time(s)')
plt.ylabel('Frequency(Hz)')

F1 = df.F1_Hz
F2 = df.F2_Hz
F3 = df.F3_Hz.astype(float)
F4 = df.F4_Hz.astype(float)

plt.plot(time, F1, 'ro', label='F1(Hz)')
plt.plot(time, F2, 'bo', label='F2(Hz)')
plt.plot(time, F3, 'go', label='F3(Hz)')
plt.plot(time, F4, 'yo', label='F4(Hz)')

plt.legend(loc='best', ncol=2, fontsize=14, frameon=True, shadow=True)
plt.show()


```
